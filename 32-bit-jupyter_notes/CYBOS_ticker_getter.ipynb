## KOSPI, KOSDAQ 리스트 반환

`value = object.GetStockMarketKind ( code ) `

code 에해당하는소속부를반환한다. 

code : 주식코드

반환값 : 소속부
```
typedefenum{
[helpstring("구분없음")]CPC_MARKET_NULL= 0, 
[helpstring("거래소")]   CPC_MARKET_KOSPI= 1, 
[helpstring("코스닥")]   CPC_MARKET_KOSDAQ= 2, 
[helpstring("K-OTC")] CPC_MARKET_FREEBOARD= 3, 
[helpstring("KRX")]       CPC_MARKET_KRX= 4,
[helpstring("KONEX")] CPC_MARKET_KONEX= 5,
}CPE_MARKET_KIND; 
```

`value = object.GetStockListByMarket(CPE_MARKET_KIND code )`

시장구분에따른주식종목배열을반환하다

반환값: 입력한시장구분(`CPE_MARKET_KIND`)에해당하는종목리스트(배열)


````python
instCpCodeMgr = win32com.client.Dispatch('CpUtil.CpCodeMgr')
kospi_code_list = instCpCodeMgr.GetStockListByMarket(1)
kosdaq_code_list = instCpCodeMgr.GetStockListByMarket(2)
````

`value = object.GetStockSectionKind( code )`

code 에해당하는부구분코드를반환한다

code : 주식코드

반환값 : 부구분코드

```
typedefenum

{

   [helpstring("주권")]   CPC_KSE_SECTION_KIND_ST   = 1,

  [helpstring("사회간접자본투융자회사")]CPC_KSE_SECTION_KIND_IF = 5,

  [helpstring("주식예탁증서")]   CPC_KSE_SECTION_KIND_DR    = 6,

  [helpstring("상장지수펀드(ETF)")]CPC_KSE_SECTION_KIND_ETF = 10,

  [helpstring("옵션")]      CPC_KSE_SECTION_KIND_OP    = 15,    

} CPE_KSE_SECTION_KIND;
```

````python
# kospi

kospi = []
for code in kospi_code_list:   
    name = instCpCodeMgr.CodeToName(code)
    stock_control_kind = instCpCodeMgr.GetStockControlKind(code) # 감리구분
    stock_supervision_kind = instCpCodeMgr.GetStockSupervisionKind(code) # 관리구분 
    stock_status_kind = instCpCodeMgr.GetStockStatusKind(code) # 주식상태
    stock_kospi200_kind = instCpCodeMgr.GetStockKospi200Kind(code) # KOSPI200 종목여부
    stock_section_kind = instCpCodeMgr.GetStockSectionKind(code) # 부구분코드
    
    row_dic = {}
    row_dic['code'] = code
    row_dic['name'] = name
    row_dic['stock_control_kind'] = stock_control_kind
    row_dic['stock_supervision_kind'] = stock_supervision_kind
    row_dic['stock_status_kind'] = stock_status_kind
    row_dic['stock_kospi200_kind'] = stock_kospi200_kind
    row_dic['stock_section_kind'] = stock_section_kind
    
    kospi.append(row_dic)
````

````python
kosdaq = []
for code in kosdaq_code_list:
    name = instCpCodeMgr.CodeToName(code)
    stock_control_kind = instCpCodeMgr.GetStockControlKind(code) # 감리구분
    stock_supervision_kind = instCpCodeMgr.GetStockSupervisionKind(code) # 관리구분 
    stock_status_kind = instCpCodeMgr.GetStockStatusKind(code) # 주식상태
    stock_kospi200_kind = instCpCodeMgr.GetStockKospi200Kind(code) # KOSPI200 종목여부
    stock_section_kind = instCpCodeMgr.GetStockSectionKind(code) # 부구분코드
    
    row_dic = {}
    row_dic['code'] = code
    row_dic['name'] = name
    row_dic['stock_control_kind'] = stock_control_kind
    row_dic['stock_supervision_kind'] = stock_supervision_kind
    row_dic['stock_status_kind'] = stock_status_kind
    row_dic['stock_kospi200_kind'] = stock_kospi200_kind
    row_dic['stock_section_kind'] = stock_section_kind
    
    kosdaq.append(row_dic)
````

````python
kospi_df = pd.DataFrame(kospi)
kospi_df.head()
````

````python
kosdaq_df = pd.DataFrame(kosdaq)
kosdaq_df.head()
````
